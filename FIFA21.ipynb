# Using K-Means Clustering Algorithm

```python

numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns


numeric_columns.tolist()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt



# Select relevant features and drop missing values
features = ['OVA', 'POT', 'Acceleration', 'Sprint Speed', 'Shot Power']
subset_data = data[features].dropna()

# Scale the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(subset_data)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4, random_state=30)
subset_data['Cluster'] = kmeans.fit_predict(scaled_data)

# Display the first 10 rows
subset_data.head(10)
```

```python
print(subset_data['Cluster'].value_counts())
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(subset_data['OVA'], subset_data['POT'], c=subset_data['Cluster'], cmap='viridis', alpha=0.8)
plt.xlabel('OVA')
plt.ylabel('POT')
plt.title('Cluster Visualization')
plt.colorbar(label='Cluster')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=30)
subset_data['Cluster'] = kmeans.fit_predict(scaled_data)
```

```python
wcss = []
for i in range(1, 11):  # Test 1 to 10 clusters
    kmeans = KMeans(n_clusters=i, random_state=30)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal Clusters')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=30, n_init=20)
subset_data['Cluster'] = kmeans.fit_predict(scaled_data)
# Add player names to the dataset after clustering
subset_data['Name'] = data['Name'].iloc[subset_data.index]  # Add 'Name' column from original dataset

# Display the updated dataset with player names and cluster labels
subset_data[['Name', 'OVA', 'POT', 'Acceleration', 'Sprint Speed', 'Shot Power','Cluster']].head(10).sort_values(by='Cluster')

```

```python
# Scatter plot to visualize clusters based on OVA and POT
plt.figure(figsize=(8, 6))
for cluster in subset_data['Cluster'].unique():
    cluster_data = subset_data[subset_data['Cluster'] == cluster]
    plt.scatter(cluster_data['OVA'], cluster_data['POT'], label=f'Cluster {cluster}', alpha=0.7)

# Add labels and legend
plt.xlabel('Overall Rating (OVA)')
plt.ylabel('Potential (POT)')
plt.title('K-Means Clustering: OVA vs POT')
plt.legend()
plt.show()
```
